Para las imágenes suministradas:

    -Implementar un extractor de carácteristicas LBP básico (sin uniformidad, 8 vecinos)
    -Realizar el histograma de carácterísticas LBP de la imágen
    -Comparar los histogramas

```python
%matplotlib inline

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
potencia = np.array([128, 64, 32, 16, 8, 4, 2, 1])
```

```python
#Imagen 1
imagen1 = cv.imread('Imagen1.png')
imagen1_RGB = cv.cvtColor(imgagen1,cv.COLOR_BGR2RGB)
gray = cv.cvtColor(imgagen1, cv.COLOR_BGR2GRAY)
gray = np.pad(gray, pad_width = 1, mode = 'edge')
gray_LRB1 = np.zeros_like(gray)
for i in range(1, gray.shape[0]-1):
    for j in range(1, gray.shape[1]-1):
        ventana = (np.append(np.append(np.append(gray[i-1,j-1:j+2],gray[i,j+1]),np.flip(gray[i+1,j-1:j+2])),gray[i,j-1]))
        gray_LRB1[i-1,j-1] = np.sum((ventana >= gray[i,j])*potencia)
figura = plt.figure(figsize = (10,10))
ax1 = figura.add_subplot(1, 2, 1)
ax1.imshow(imgagen1_RGB)
ax1.set_title('Imagen1'.format(imagen1))
ax2 = figura.add_subplot(1, 2, 2)
ax2.hist(np.ravel(gray_LRB1), bins = 256, color = 'g')
ax2.set_title('Imagen1: Histograma LBP'.format(imagen1))
plt.show()
```
