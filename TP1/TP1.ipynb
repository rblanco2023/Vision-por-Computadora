## Coordenadas cromáticas

```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/home/raul/Documentos/UBA/Visión por computadora 1/Clase 2/TP1/CoordCrom_1.png')
img2 = cv.imread('/home/raul/Documentos/UBA/Visión por computadora 1/Clase 2/TP1/CoordCrom_2.png')
img3 = cv.imread('/home/raul/Documentos/UBA/Visión por computadora 1/Clase 2/TP1/CoordCrom_3.png')
```

```python
b, g, r = cv.split(img)
b_integer=b.astype(int)
g_integer=g.astype(int)
r_integer=r.astype(int)
b2, g2, r2 = cv.split(img2)
b2_integer=b2.astype(int)
g2_integer=g2.astype(int)
r2_integer=r2.astype(int)
b3, g3, r3 = cv.split(img3)
b3_integer=b3.astype(int)
g3_integer=g3.astype(int)
r3_integer=r3.astype(int)

den=r_integer+g_integer+b_integer
den[den==0]=1
rgb=cv.merge((r, g, b))
croma=cv.merge([r_integer/den,g_integer/den,b_integer/den])

den2=r2_integer+g2_integer+b2_integer
den2[den2==0]=1
rgb2=cv.merge((r2, g2, b2))
croma2=cv.merge([r2_integer/den2,g2_integer/den2,b2_integer/den2])

den3=r3_integer+g3_integer+b3_integer
den3[den3==0]=1
rgb3=cv.merge((r3, g3, b3))
croma3=cv.merge([r3_integer/den3,g3_integer/den3,b3_integer/den3])

figura=plt.figure()
axis_1=plt.subplot(3,2,1)
axis_1.imshow(rgb)
axis_1.set_title('Imagen 1 original')
axis_2=plt.subplot(3,2,2)
axis_2.imshow(croma)
axis_2.set_title('Imagen 1 cromática')

axis_3=plt.subplot(3,2,3)
axis_3.imshow(rgb2)
axis_3.set_title('Imagen 2 original')
axis_4=plt.subplot(3,2,4)
axis_4.imshow(croma2)
axis_4.set_title('Imagen 2 cromática')

axis_5=plt.subplot(3,2,5)
axis_5.imshow(rgb3)
axis_5.set_title('Imagen 3 original')
axis_6=plt.subplot(3,2,6)
axis_6.imshow(croma3)
axis_6.set_title('Imagen 3 cromática')

plt.show()
```

```python
imagen=cv.imread('/home/raul/Documentos/UBA/Visión por computadora 1/Clase 2/TP1/WP_B.png')
b,g,r = cv.split(imagen)
r_max = np.amax(r)
g_max = np.amax(g)
b_max = np.amax(b)
s_r = 255/r_max
s_g = 255/g_max
s_b = 255/b_max
figura=plt.figure()
axis1=plt.subplot(1,2,1)
axis1.imshow(cv.merge([r,g,b]))
axis1.set_title('B original')
axis2=plt.subplot(1,2,2)
axis2.imshow(cv.merge([(s_r*r).astype('uint8'),(s_g*g).astype('uint8'),(s_b*b).astype('uint8')]))
axis2.set_title('B White patch')
plt.show()
```
